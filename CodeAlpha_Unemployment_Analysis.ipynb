```python
# ==========================================================
# CODEALPHA TASK 2
# UNEMPLOYMENT ANALYSIS WITH PYTHON
# ==========================================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files

# ==========================================================
# Upload Dataset
# ==========================================================

print("Upload both CSV files")
uploaded = files.upload()

# ==========================================================
# Load Datasets
# ==========================================================

df = pd.read_csv("Unemployment in India.csv")
covid_df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

# ==========================================================
# Clean Column Names
# ==========================================================

df.columns = df.columns.str.strip()
covid_df.columns = covid_df.columns.str.strip()

# ==========================================================
# Convert Date
# ==========================================================

df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
covid_df["Date"] = pd.to_datetime(covid_df["Date"], dayfirst=True)

# ==========================================================
# Basic Information
# ==========================================================

print("\n========== DATASET ==========")
print(df.head())

print("\nShape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values")
print(df.isnull().sum())

print("\nStatistics")
print(df.describe())

# ==========================================================
# Drop Missing Values
# ==========================================================

df = df.dropna()

# ==========================================================
# Average Unemployment by Region
# ==========================================================

region_unemployment = (
    df.groupby("Region")["Estimated Unemployment Rate (%)"]
      .mean()
      .sort_values(ascending=False)
)

print("\nAverage Unemployment Rate by Region")
print(region_unemployment)

# ==========================================================
# Graph 1
# ==========================================================

plt.figure(figsize=(12,6))
region_unemployment.plot(kind="bar")
plt.title("Average Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# ==========================================================
# Urban vs Rural
# ==========================================================

area = (
    df.groupby("Area")["Estimated Unemployment Rate (%)"]
      .mean()
)

print("\nUrban vs Rural")
print(area)

plt.figure(figsize=(5,5))
area.plot(kind="bar")
plt.title("Average Unemployment by Area")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()

# ==========================================================
# Monthly Trend
# ==========================================================

monthly = (
    df.groupby(df["Date"].dt.month)["Estimated Unemployment Rate (%)"]
      .mean()
)

plt.figure(figsize=(8,5))
monthly.plot(marker="o")
plt.title("Monthly Average Unemployment")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()

# ==========================================================
# COVID Dataset
# ==========================================================

covid_df = covid_df.dropna()

covid_region = (
    covid_df.groupby("Region")["Estimated Unemployment Rate (%)"]
            .mean()
            .sort_values(ascending=False)
)

print("\nCOVID Period Unemployment")
print(covid_region)

plt.figure(figsize=(12,6))
covid_region.plot(kind="bar", color="orange")
plt.title("COVID-19 Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# ==========================================================
# Top 10 Highest Unemployment Regions
# ==========================================================

top10 = covid_region.head(10)

plt.figure(figsize=(8,5))
top10.plot(kind="bar", color="red")
plt.title("Top 10 Regions with Highest Unemployment")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()

# ==========================================================
# Conclusion
# ==========================================================

print("\n========== PROJECT CONCLUSION ==========")

print("""
1. The unemployment rate varies significantly across regions.

2. COVID-19 caused a noticeable increase in unemployment in many states.

3. Rural and Urban areas have different unemployment patterns.

4. Monthly unemployment trends show fluctuations throughout the year.

5. These insights can help policymakers identify regions requiring employment support.
""")

print("\nProject Completed Successfully!")
```
